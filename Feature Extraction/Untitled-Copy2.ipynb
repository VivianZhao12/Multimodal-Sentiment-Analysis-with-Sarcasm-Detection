```python
import pandas as pd
```

```python
audio = pd.read_csv('../datasets/audio_sarcasm_dataset.csv')
audio.columns
```

```python
# Assuming df is your DataFrame
audio.rename(columns={
    'AUDIO_1': 'PCA_MFCC_1',
    'AUDIO_2': 'PCA_MFCC_2',
    'AUDIO_3': 'PCA_MFCC_3',
    'AUDIO_4': 'PCA_MFCC_4',
    'AUDIO_5': 'PCA_MFCC_5',
    'AUDIO_6': 'PCA_MFCC_6',
    'AUDIO_7': 'PCA_MFCC_7',
    'AUDIO_8': 'PCA_MFCC_8'
}, inplace=True)
```

```python
audio = audio.drop(columns = ['energy', 'loudness'])
```

```python
pd.read_csv('../datasets/text_sarcasm_dataset.csv').columns
```

```python
text = pd.read_csv('../datasets/text_sarcasm_dataset.csv')

# Assuming df is your DataFrame
text.rename(columns={
    f'TEXT_{i}': f'PCA_W2V_{i}' for i in range(1, 61)  # This creates mappings from TEXT_1 to TEXT_50
}, inplace=True)

text.columns
```

```python
text = text[['id', 'PCA_W2V_1', 'PCA_W2V_2', 'PCA_W2V_3', 'PCA_W2V_4',
       'PCA_W2V_5', 'PCA_W2V_6', 'PCA_W2V_7', 'PCA_W2V_8', 'PCA_W2V_9',
       'PCA_W2V_10', 'PCA_W2V_11', 'PCA_W2V_12', 'PCA_W2V_13', 'PCA_W2V_14',
       'PCA_W2V_15', 'PCA_W2V_16', 'PCA_W2V_17', 'PCA_W2V_18', 'PCA_W2V_19',
       'PCA_W2V_20', 'PCA_W2V_21', 'PCA_W2V_22', 'PCA_W2V_23', 'PCA_W2V_24',
       'PCA_W2V_25', 'PCA_W2V_26', 'PCA_W2V_27', 'PCA_W2V_28', 'PCA_W2V_29',
       'PCA_W2V_30', 'PCA_W2V_31', 'PCA_W2V_32', 'PCA_W2V_33', 'PCA_W2V_34',
       'PCA_W2V_35', 'PCA_W2V_36', 'PCA_W2V_37', 'PCA_W2V_38', 'PCA_W2V_39',
       'PCA_W2V_40', 'PCA_W2V_41', 'PCA_W2V_42', 'PCA_W2V_43', 'PCA_W2V_44',
       'PCA_W2V_45', 'PCA_W2V_46', 'PCA_W2V_47', 'PCA_W2V_48', 'PCA_W2V_49',
       'PCA_W2V_50']]
```

```python
merged_df = audio.merge(text, on='id', suffixes=('_audio', '_text'))

# Filter out the _text columns that have a corresponding _audio column
for col in merged_df.columns:
    if col.endswith('_text'):
        audio_col = col.replace('_text', '_audio')
        if audio_col in merged_df.columns:
            merged_df.drop(col, axis=1, inplace=True)  # Drop the text column if the audio column exists
        else:
            # Rename back to original if no corresponding audio column exists
            merged_df.rename(columns={col: col.replace('_text', '')}, inplace=True)

# Optionally, rename columns to remove the suffix from audio columns where no overlap occurred
for col in merged_df.columns:
    if col.endswith('_audio'):
        merged_df.rename(columns={col: col.replace('_audio', '')}, inplace=True)

# Now merged_df contains all columns from 'audio' and only non-overlapping columns from 'text'
```
